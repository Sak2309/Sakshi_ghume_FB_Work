# 1.line chart : used 1.points are linaer or in one line or cured 

**to plot a line x,y**

```python
plt.plot(x,y)
```

**to plot dots.** 

```python
plt.plot(x,y,marker ='p') #'o'
```

**linestyle**

```python
plt.plot(x,y,marker ='o',linestyle='dashed')
```

**figure size**

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,marker ='p')
```

**linec0lor**

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,marker ='o',linestyle='dashed',color = 'red',markerfacecolor = 'black')
```

**to label the fig x,y**

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,marker ='o',linestyle='dashed',color = 'red',markerfacecolor = 'black')
plt.xlabel('numbers')
plt.ylabel('square')
```

**plt.plot :gives the object detils in text .u want remove the we used plt.show() method**

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,marker ='o',linestyle='dashed',color = 'red',markerfacecolor = 'black')
plt.xlabel('numbers')
plt.ylabel('square')
plt.show()
```

**to save fig or ghraph in pic format**

```python
plt.figure(figsize=(5,5))
plt.plot(x,y,marker ='o',linestyle='dashed',color = 'red',markerfacecolor = 'black')
plt.xlabel('numbers')
plt.ylabel('square')
plt.savefig("linear_char_fig.jpg")
```
